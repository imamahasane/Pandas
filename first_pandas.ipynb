# Group by and Aggregating

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Flavors.csv')

df
```

```python
df.groupby('Base Flavor').mean('Flavor Rating', 'Texture Rating', 'Total Rating')
```

### or

```python
group_by_frame = df.groupby('Base Flavor').agg({
    'Flavor Rating': 'mean',
    'Texture Rating': 'mean',
    'Total Rating': 'mean',
    # 'Liked': 'count',  # Counting how many entries per group
    # 'Flavor': 'first'  # Taking the first entry for the 'Flavor' column
})
```

```python
group_by_frame
```

```python
 df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby(['Base Flavor', 'Liked']).mean('Flavor Rating', 'Texture Rating', 'Total Rating')
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby('Base Flavor').describe()
```

# Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv('LOTR.csv')
df1
```

```python
df2 = pd.read_csv('LOTR 2.csv')
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'] )
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how ='cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join = 'inner')
```

```python
pd.concat([df1, df2], join = 'outer')
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```

```python
df1
```

```python
df2
```

```python
# The error occurs because the .append() method is no longer supported in recent versions of pandas (as of version 2.0.0). 
# Instead, you can use the pd.concat() method to combine DataFrames.

# df1.append(df2)

# Combine df1 and df2 using pd.concat
df_combined = pd.concat([df1, df2], ignore_index=True)
df_combined
```
